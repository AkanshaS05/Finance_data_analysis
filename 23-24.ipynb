```python
df = pd.read_csv('23-24Union_Budget.csv', encoding='latin1')
df
```

```python
# Standardize column names
df.columns = ['Ministry', 'Demand', 'Budget_2023_Revenue', 'Budget_2023_Capital', 'Budget_2023_Total']

# Remove "Grand Total" row
df = df[df['Ministry'] != 'Grand Total']
```

```python

# Handle missing values
df.fillna(0, inplace=True)
```

```python

# Convert budget columns to numeric
numeric_cols = ['Budget_2023_Revenue', 'Budget_2023_Capital', 'Budget_2023_Total']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python

# Add derived metrics
total_budget = df['Budget_2023_Total'].sum()
df['Percentage_of_Total'] = (df['Budget_2023_Total'] / total_budget) * 100
df['Budget_Category'] = pd.cut(df['Budget_2023_Total'], 
                               bins=[0, 5000, 20000, df['Budget_2023_Total'].max()], 
                               labels=['Low', 'Medium', 'High'])


```

```python
df
```

```python
#remove any rows with missing values
df.dropna(inplace=True)
```

```python
# Save cleaned data
df.to_csv('cleaned_budget_data.csv', index=False)
```
